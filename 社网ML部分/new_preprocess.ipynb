```python
usr_info['follow_ratio'] = usr_info['粉丝数'] / usr_info['关注数']

data_norm = pd.read_csv(r'new/norm/weibo.csv',encoding='UTF-8')
data_fake = pd.read_csv(r'new/spam/weibo.csv',encoding='UTF-8')
frames=[data_norm,data_fake]
dataset = pd.concat(frames,ignore_index=True)
dataset
```

```python

# 被转评赞的总数，平均数
usr_info['like_count']=0
usr_info['repost_count']=0
usr_info['comment_count']=0
usr_info['tweet_count']=0
usr_info['ave_like']=0.
usr_info['ave_repost']=0.
usr_info['ave_comment']=0.

usr_info['att_count']=0
usr_info['jing_count']=0
usr_info['zhuanfa_count']=0
usr_info['url_count']=0
usr_info['ave_att']=0.
usr_info['ave_jing']=0.
usr_info['ave_zhuanfa']=0.
usr_info['ave_url']=0.


for i,usr_id in enumerate(usr_info.id):
    # usr_tweet = []
    tweet_num=0
    usr_like=0
    usr_repost=0
    usr_comment=0
    for j,tweet_id in enumerate(dataset['用户id']):
        if usr_id == tweet_id:
            # print(dataset['内容'][j:j+1].to_string())
            if '#' in dataset['内容'][j:j+1].to_string():
                usr_info.loc[i,'jing_count'] +=1
            if '@' in dataset['内容'][j:j+1].to_string():
                usr_info.loc[i,'att_count'] +=1
            if 'http' in dataset['内容'][j:j+1].to_string():
                usr_info.loc[i,'url_count'] +=1
            if '转发微博' in dataset['内容'][j:j+1].to_string():
                usr_info.loc[i,'zhuanfa_count'] +=1
            # usr_tweet.append(dataset['内容'][j:j+1]) #找该用户发的所有的推文
            tweet_num+=1
            # print(dataset['点赞数'][i:i+1])
            usr_like+=int(dataset['点赞数'][j:j+1].tolist()[0])
            usr_comment+=int(dataset['评论数'][j:j+1].tolist()[0])
            usr_repost+=int(dataset['转发数'][j:j+1].tolist()[0])
    # abort()
    usr_info.loc[i,'like_count'] = usr_like
    usr_info.loc[i,'repost_count'] = usr_repost
    usr_info.loc[i,'comment_count'] = usr_comment
    usr_info.loc[i,'tweet_count'] = tweet_num
    if tweet_num>0:
        usr_info.loc[i,'ave_like'] = usr_like/tweet_num
        usr_info.loc[i,'ave_comment'] = usr_comment/tweet_num
        usr_info.loc[i,'ave_repost'] = usr_repost/tweet_num
        usr_info.loc[i,'ave_jing'] = usr_info.loc[i,'jing_count']/tweet_num
        usr_info.loc[i,'ave_att'] = usr_info.loc[i,'att_count']/tweet_num
        usr_info.loc[i,'ave_url'] = usr_info.loc[i,'url_count']/tweet_num
        usr_info.loc[i,'ave_zhuanfa'] = usr_info.loc[i,'zhuanfa_count']/tweet_num
```

```python
usr_info = pd.read_csv(r'new/usr_info.csv',encoding='UTF-8')
# usr_info = usr_info.drop(["_id",'nick_name','vip_level'], axis = 1) #把user id等无关变量去掉

real_rela = pd.read_csv(r'new/norm/Edges.csv')
fake_rela = pd.read_csv(r'new/spam/Edges.csv')

frames = [real_rela, fake_rela]
rela = pd.concat(frames,ignore_index=False)


# usr_info
usr_info['fan_fan_ave']=0#粉丝的粉丝平均数
usr_info['followed_fan_ave']=0#关注者的粉丝平均数
usr_info['mutual']=0#互关数
usr_info['sina_helper_fan']=False
for i ,row in enumerate(usr_info.id):
    fan_data = rela.loc[rela['target'] == row]
    follower_data = rela.loc[rela['source'] == row]
    fans_num=0
    fans_tot = 0
    follows_num=0
    follows_tot = 0
    mutual = 0
    if not fan_data.empty:
        for j, fanner_id in enumerate(fan_data.source):
            fan_info = usr_info.loc[usr_info.id == fanner_id]
            if not fan_info.empty and fan_info['粉丝数'].values[0]<1e7:
                fans_num += fan_info['粉丝数'].values[0]
                fans_tot+=1
            if not fan_info.empty and fan_info['粉丝数'].values[0]>1e7:
                usr_info.loc[i,'sina_helper_fan']=True
        if fans_tot!=0:
            usr_info.loc[i,'followed_fan_ave'] = fans_num / fans_tot

    if not follower_data.empty:
        for j, follower_id in enumerate(follower_data.target):
            follower_info = usr_info.loc[usr_info.id == follower_id]
            if not follower_info.empty:
                follows_num += follower_info['粉丝数'].values[0]
                follows_tot+=1
        if follows_tot!=0:
            usr_info.loc[i,'fan_fan_ave'] = follows_num / follows_tot
    if not fan_data.empty and not follower_data.empty:
        for j, fanner_id in enumerate(fan_data.source):
            fan_info = follower_data.loc[follower_data.target == fanner_id]#是否互关
            if not fan_info.empty:
                mutual +=1
        usr_info.loc[i,'mutual'] = mutual
```

```python
usr_info = pd.read_csv(r'new/new_usr_info.csv',encoding='UTF-8')
from collections import defaultdict
# 将主题作为特征
weibo_info = pd.read_csv(r'new/weibo_topic.csv',encoding='utf-8')
usr_info['theme_num']=0 #微博的主题个数
usr_info['most_theme_pct']=0 #微博最多主题的比例
usr_info['most_theme']='' #最多主题编号

for i ,row in enumerate(usr_info.id):
    theme_num=0
    most_theme_pct=0
    most_theme=''
    topicDict=defaultdict(lambda: 0)
    usr_weibo = weibo_info.loc[weibo_info['用户id'] == row]
    for j, single_topic in enumerate(usr_weibo.topic):
        topicDict[str(single_topic)]+=1
    theme_num = len(topicDict.keys())
    if theme_num>0:
        most_theme = max(topicDict, key=topicDict.get)
        most_theme_pct = max(topicDict.values()) /sum(topicDict.values())
    usr_info.loc[i,'theme_num'] = theme_num
    usr_info.loc[i,'most_theme_pct'] = most_theme_pct
    usr_info.loc[i,'most_theme'] = most_theme

from sklearn.preprocessing import OneHotEncoder    
encoder = OneHotEncoder(sparse=False)# ordinal 指标转换成8个0-1变量
most_theme_numeric = encoder.fit_transform(usr_info[['most_theme']])
# most_theme_numeric[:5]
feature_names = encoder.get_feature_names_out(usr_info[['most_theme']].columns)
most_theme_numeric = pd.DataFrame(most_theme_numeric, columns=feature_names)
new_usr_info = pd.concat([usr_info, most_theme_numeric], axis=1)
new_usr_info = new_usr_info.drop(['most_theme'],axis=1)
```
