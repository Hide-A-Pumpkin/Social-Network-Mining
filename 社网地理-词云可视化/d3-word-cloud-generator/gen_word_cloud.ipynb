```python
import numpy as np
import pandas as pd
import csv
import math

with open("freq_real.csv",'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    freq_real = list(reader)
```

```python
wordList=''
freq_real
```

```python
for i in range(200):
    wordList += (freq_real[i][0]+' ')*int(math.sqrt(int(freq_real[i][1])))
```

```python
with open('freq_real.txt','w') as file:
    file.write(wordList)
```
